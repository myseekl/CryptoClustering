```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")


# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

# Choose columns from dataframe to scale
df_market_data_columns_to_scale = df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d",	"price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]]

# Fit & transform the data
scaled_market_data = scaler.fit_transform(df_market_data)
print(scaled_market_data)
```

```python
# Create a DataFrame with the scaled data
scaled_market_data_df = pd.DataFrame(scaled_market_data, columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d",	"price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"])


# Copy the crypto names from the original DataFrame

# Reset the index to add 'coin_id' as a column
df_market_data.reset_index(inplace=True)

scaled_market_data_df['coin_id'] = df_market_data['coin_id']


# Set the coin_id column as index
scaled_market_data_df.set_index("coin_id", inplace=True)

# Display the scaled DataFrame
print(scaled_market_data_df)
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Define the range of k values
k_values = range(1, 12)  # Adjusted to include 1 to 11

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_market_data_df)
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.

df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4
